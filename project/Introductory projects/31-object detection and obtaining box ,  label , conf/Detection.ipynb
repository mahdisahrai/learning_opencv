### importing

```
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

### obtaining box,label,conf for any object

```
img = cv2.imread('/content/sample_data/car.jpg')
box,label,conf = cv.detect_common_objects(img)
output = draw_bbox(img,box,label,conf)
plt.imshow(output)
plt.show()
```

```
print('number of cars: ' + str(label.count('car')))
```
